1) Escreva um programa, usando Python, Matplotlib e bibliotecas associadas, para criar
um gráfico de setores das medalhas olímpicas obtidas pelos cinco países mais bem sucedidos
na Olímpiada de verão de 2016. Utilize os seguintes dados.

Alemanha, 17

China, 26

Estados Unidos, 46

Grã-Bretanha, 27

Rússia, 19 

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(14, 7), subplot_kw=dict(aspect="equal"))

medalhas = ["19 Rússia",
          "17 Alemanha",
          "46 EstadosUnidos",
          "27 Grã-Bretanha",
          "26 China"]

data = [float(x.split()[0]) for x in medalhas]
Países = [x.split()[-1] for x in medalhas]


def func(pct, allvals):
    absolute = int(np.round(pct/100.*np.sum(allvals)))
    return "{:.1f}%\n({:d} medalhas)".format(pct, absolute)


wedges, texts, autotexts = ax.pie(data, autopct=lambda pct: func(pct, data),
                                  textprops=dict(color="w"))

ax.legend(wedges, Países,
          title="Países",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=10, weight="bold")

ax.set_title("Medalhas de ouro dos 5 melhores países\n na Olímpiada de verão de 2016")

plt.show()
```

2) Escreva um programa, usando Python, Matplotlib e bibliotecas associadas, para criar
um gráfico de barras mostrando a popularidade das principais linguagens de programação
em 2017. Utilize os seguintes dados.

Linguagens de programação: Java, Python, PHP, JavaScript, C#, C++

      Popularidade:              22.2,   17.6,   8.8,    8.0,    7.7,  6.7



```python
import matplotlib.pyplot as plt
import numpy as np



labels = ['                   Java', '                   Python', '                   PHP', '                   Javascript', '                   C#', '                   C++' ]
linguagens = [22.2, 17.6, 8.8, 8.0, 7.7, 6.7]

x = labels

width = 0.70  

fig, ax = plt.subplots(figsize=(10,5))

rects1 = ax.bar(x, linguagens, width, color="green", align='edge')

ax.set_ylim(0.,25)
ax.set_yticks(np.linspace(0, 25, 6, endpoint=True))

ax.set_ylabel('Popularidade')
ax.set_xlabel('Linguagens de programação')
ax.set_title('Popularidade das linguagens de programação em 2017')


plt.grid(color = 'blue', linestyle = '--', linewidth = 0.5, alpha=1.0)

plt.minorticks_on()

ax.grid(True, which='both')
ax.grid(which='minor', alpha=0.3)



fig.tight_layout()

plt.show()
```
